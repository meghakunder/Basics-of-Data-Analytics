# Basics of Data Analytics

 1.Define the role of a Data Analyst in your own words. What value do they bring to an organization?

       

*  A Data Analyst collects, cleans, and studies data to help organizations make better decisions.They find patterns, create reports, and explain what the data means in simple terms.They convert raw data into meaningful insights that support business growth and efficiency.

3.Write the end-to-end analytics workflow in the correct order and briefly explain each step.



*   Data Collection – Gathering data from various sources like databases, files, APIs, etc.
*   Data Cleaning – Removing errors, duplicates, and missing values to make data accurate.
*   Data Analysis – Studying the data using tools and statistics to find patterns or insights.
*   Data Visualization – Creating charts and dashboards to show findings
clearly.
*   Decision Making – Using insights from the data to take business actions or solve problems.

4.What is Prompt Engineering, and explain any two types of prompts used in Generative AI?
     
->   Prompt Engineering means creating effective instructions/ prompts to get the best results from AI tools.



Two Types of Prompts Used in Generative AI

Instruction Prompt:
This type of prompt gives a clear command or question to the AI to perform a specific task.
Example: Summarize this paragraph in one sentence.

Example-Based Prompt:
This prompt includes one or more examples to help the AI understand the pattern or style before giving its response.

5.What are the key differences between a Business Analyst and a Data Analyst in terms of roles,
responsibilities, and focus?
  

###    Business Analyst

*   Role:Focuses on understanding business needs and finding solutions to improve processes
*    Responsibilities:Gathers requirements, interacts with stakeholders, creates business reports, and suggests improvements.

*    Focus:Business operations, strategy, and process improvement.Helps make business and strategic decisions.
*   Tools:Excel, Visio, Jira, MS Office, PowerPoint.

###Data Analyst


*   Role:Focuses on collecting, cleaning, and analyzing data to find insights.
*   Responsibilities:Extracts and prepares data, performs analysis, builds dashboards, and presents data-driven findings.
*  Focus:Data trends, patterns, and performance metrics.
*   Tools:SQL, Excel, Python, Power BI, Tableau.

6.Explain any three AI-powered ETL (Extract, Transform, Load) tools and how they are used in analytics.


*   Airbyte:
Airbyte is an open-source ETL tool that collects data from different sources and loads it into data warehouses.
It uses AI to automate data connections and keep data updated for faster analysis.


*   Azure Data Factory:
Azure Data Factory is a cloud-based ETL tool that uses AI to automate data movement and transformation.
It helps combine data from multiple systems for reporting and analytics.

*   AWS Glue:
AWS Glue is an AI-powered ETL service by Amazon that automatically discovers, cleans, and organizes data.
It prepares data quickly for dashboards and business insights.

8.Explain the evolution of analytics, highlighting the different stages from Descriptive to Generative AI.
    

*   Descriptive Analytics – Explains what happened .
*  Diagnostic Analytics – Explains why it happened -finds root causes.
*   Predictive Analytics – Predicts what is likely to happen next.
*   Prescriptive Analytics – Suggests what actions to take .How can we make it happen?
*   Generative AI Analytics – Uses AI models  to create new content, automate reports, and generate insights from data in natural language.
